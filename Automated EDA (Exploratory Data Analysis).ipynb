```python
import numpy as np 
import pandas as pd 
import matplotlib.pylab as plt
import seaborn as sns 
```

```python
df = pd.read_csv('International_Report_Departures.csv')
```

```python
new_data_with_noNAN=df[~df.carrier.isna()]
new_data_with_noNAN
```

```python
new_data_with_noNAN.info(),new_data_with_noNAN.isna().mean()
```

```python
the_most_year= new_data_with_noNAN.Year.sort_index().value_counts()
the_most_year
```

```python
plt.figure(figsize=(30,20))
plt.style.use("ggplot")
plt.bar(the_most_year.index , the_most_year.values)
plt.xlabel("year")
plt.ylabel('frequancey')
plt.xticks(rotation=90)
```

```python
most_carrer = new_data_with_noNAN.carrier.sort_values().value_counts()[0:10]
```

```python
most_carrer
```

```python
plt.figure(figsize=(30,30))
fig ,ax= plt.subplots(1,2)
ax[0].bar(most_carrer.index,most_carrer.values)
ax[0].set_xlabel("carrer")
ax[0].set_ylabel("grequancey")
ax[1].pie(most_carrer ,labels=most_carrer.index ,autopct='%1.2f%%', explode=[0.03,0,0.01,0,0.03,0.003,0,0,0,.03])
```
